Seaborn data visualization library


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({
                    'No': [1,2,3,4,5,6],
                   'Country':['India','UK','Usa','Pak','AFG','Bangladesh'],
                   'Inflamation Rate': [8,9,10,11,10,8],
                   'GDP': [5,4,3,2,8,9],
                   'Sales':[1000,4000,6000,8000,9000,6000]
                   })
df
```

```python
# importing seaborn library
import seaborn as sns
import pandas as pd 
import numpy as np
sns.lineplot(data = df,x = df['Country'],y =df['Sales'],linestyle = '--',color ='g',marker ='o')
plt.show()
```

```python
# barplot and lineplot with use seaborn
sns.lineplot(data =df,x=df['Country'],y =df['Sales'],linestyle = '--',color ='g',marker ='o')
plt.ylim(0,10000)

sns.barplot(data =df,x=df['Country'],y =df['Sales'],color ='lightblue')
plt.show()
```

```python
#countPlot ..occurance show , india how many time occur 
sns.countplot(data=df, x =df['Country']);
```

```python
#Displot
sns.displot(x = df['Sales'] ,kde= True) #kde means kernal distribution and curve

plt.show()
```

```python
sns.scatterplot(data =df,x=df['GDP'],y =df['Inflamation Rate'],color ='black');
```

```python
# hue parameter is numerical or categorical as well..is like tooltip or label use for data insights
sns.scatterplot(data =df,x=df['GDP'],y =df['Inflamation Rate'],color ='black',hue=df['Sales']);
```

```python
# hue parameter is numerical or categorical as well..is like tooltip or label use for data insights
sns.scatterplot(data =df,x=df['GDP'],y =df['Inflamation Rate'],color ='black',hue=df['Country']);
```

```python
#Regplot : regression plot : two numerical columns : y =mx + c**2 regression formula :
sns.regplot(data=df,x=df['GDP'],y= df['Sales'],ci=0.9,color='maroon',marker='o',label='Regression line')
plt.ylim(500,10000)
plt.xlim(0,10)
plt.legend()
plt.grid()
plt.show()
```

```python
# pair plot : o=to show all pairs in the use of data dont use for big data ...use for small data 4,5 columns
sns.pairplot(data=df)
plt.show()
```

```python
# join plot
sns.jointplot(data =df,x=df['GDP'], y = df['Sales'], hue = df['Inflamation Rate']);
```

```python
# i wnat to limit line on 7000 horizontal line and vertical line on x axis with joinplot
sns.jointplot(data =df,x=df['GDP'], y = df['Sales'], hue = df['Inflamation Rate'])
plt.axhline(7000,linestyle = '--',color = 'lightblue')
plt.axvline(6,linestyle = '--',color = 'lightblue')
plt.show()

```
